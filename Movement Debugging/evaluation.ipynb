```python
def plot_position_trajectories(df, start, columns):
    df.iloc[start:,].loc[:,columns].plot()

def plot_all_position_trajectories(df, start):
    plot_position_trajectories(df, ts, ["Pos1x", "Pos0x"])
    plot_position_trajectories(df, ts, ["Pos1y", "Pos0y"])
    plot_position_trajectories(df, ts, ["Pos1x"])
    plot_position_trajectories(df, ts, ["Pos0x"])
    plot_position_trajectories(df, ts, ["Ang10"])
    plot_position_trajectories(df, ts, ["Ang00"])
    plot_position_trajectories(df, ts, ["Ang11"])
    plot_position_trajectories(df, ts, ["Ang01"])
    plot_position_trajectories(df, ts, ["Ang11", "Ang01"])


def print_variance_and_sd(df, start, columns):
    display(df.iloc[start:,].loc[:,columns].agg({col : ['var', 'std'] for col in columns }))
```
